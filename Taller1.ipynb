```python
import math
import numpy as np
import pylab as pl
```

```python
def errorVerdadero(a,b):
    ev=a-b
    return ev

def errorAbsoluto(a,b):
    return abs(errorVerdadero(a,b))

def errorRelativo(a,b):
    if(a!=0):
        et=((a-b)/a)
    else:
        et=(((10**(-9))-b)/10**(-9))
    return et

def errorRelativoPorcentual(a,b):
    if(a!=0):
        ea=((a-b)/a)*100
    else:
        ea=(((10**(-9))-b)/(10**(-9)))*100
    return ea

def fx(x):
    r=(math.e)**x-1
    return r

def fx2(x):
    r=((math.sin(2*x)**3)+(x**2))
    return r

def fx3(x):
    r=x**4+3*(x**2)-math.cos(x)
    return r

def fpx(x):
    fo=(math.e)**x
    return fo

def fpx2(x):
    ft=6*math.sin(2*x)**2*math.cos(2*x)+2*x
    return ft

def fpx3(x):
    ftr=4*(x**3)+6*(x**2)+math.sin(x)
    return ftr
    
def a1(h,x):
    r1=(fx(x)-fx((x-h)))/h
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r1,fpx(x),errorVerdadero(fpx(x),r1),errorAbsoluto(fpx(x),r1),errorRelativo(fpx(x),r1))
    return r1
    
def a2(h,x):
    r2=(fx(x+h)-fx((x)))/h
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r2,fpx(x),errorVerdadero(fpx(x),r2),errorAbsoluto(fpx(x),r2),errorRelativo(fpx(x),r2))
    return r2

def a3(h,x):
    r3=(fx(x+h)-fx((x-h)))/(2*h)
    #print "a) %.2f  | %.5f \t | %.5f \t | %.5f \t | %.5f " %(x,r1,r2,r3,fpx(x))
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r3,fpx(x),errorVerdadero(fpx(x),r3),errorAbsoluto(fpx(x),r3),errorRelativo(fpx(x),r3))
    return r3
    

def b1(h,x):
    r1=(fx2(x)-fx2((x-h)))/h
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r1,fpx2(x),errorVerdadero(fpx2(x),r1),errorAbsoluto(fpx2(x),r1),errorRelativo(fpx2(x),r1))
    return r1

def b2(h,x):
    r2=(fx2(x+h)-fx2(x))/h
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r2,fpx2(x),errorVerdadero(fpx2(x),r2),errorAbsoluto(fpx2(x),r2),errorRelativo(fpx2(x),r2))
    return r2

def b3(h,x):
    r3=(fx2(x+h)-fx2((x-h)))/(2*h)
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r3,fpx2(x),errorVerdadero(fpx2(x),r3),errorAbsoluto(fpx2(x),r3),errorRelativo(fpx2(x),r3))
    #print "b) %.2f  | %.5f \t | %.5f \t | %.5f \t | %.5f " %(x,r1,r2,r3,fpx2(x))
    return r3

def c1(h,x):
    r1=(fx3(x)-fx3((x-h)))/h
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r1,fpx3(x),errorVerdadero(fpx3(x),r1),errorAbsoluto(fpx3(x),r1),errorRelativo(fpx3(x),r1))
    return r1

def c2(h,x):
    r2=(fx3(x+h)-fx3(x))/h
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r2,fpx3(x),errorVerdadero(fpx3(x),r2),errorAbsoluto(fpx3(x),r2),errorRelativo(fpx3(x),r2))
    return r2

def c3(h,x):
    r3=(fx3(x+h)-fx3((x-h)))/(2*h)
    print " %.2f \t | %f \t | %f \t | %f \t | %f \t | %f " %(x,r3,fpx3(x),errorVerdadero(fpx3(x),r3),errorAbsoluto(fpx3(x),r3),errorRelativo(fpx3(x),r3))
    #print "c) %.2f  | %.5f \t | %.5f \t | %.5f \t | %.5f " %(x,r1,r2,r3,fpx3(x))
    return r3


    
```
